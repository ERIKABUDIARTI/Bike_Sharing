Attribute Information (source: http://archive.ics.uci.edu/ml/datasets/Bike+Sharing+Dataset)

Both hour.csv and day.csv have the following fields, except hr which is not available in day.csv

instant: record index

dteday : date

season : season (1:spring, 2:summer, 3:fall, 4:winter)

yr : year (0: 2011, 1:2012)

mnth : month ( 1 to 12)

hr : hour (0 to 23)

holiday : weather day is holiday or not (extracted from [Web Link])

weekday : day of the week

workingday : if day is neither weekend nor holiday is 1, otherwise is 0.

weathersit :
1: Clear, Few clouds, Partly cloudy, Partly cloudy |
2: Mist + Cloudy, Mist + Broken clouds, Mist + Few clouds, Mist |
3: Light Snow, Light Rain + Thunderstorm + Scattered clouds, Light Rain + Scattered clouds |
4: Heavy Rain + Ice Pallets + Thunderstorm + Mist, Snow + Fog

temp : Normalized temperature in Celsius. The values are derived via (t-t_min)/(t_max-t_min), t_min=-8, t_max=+39 (only in hourly scale)

atemp: Normalized feeling temperature in Celsius. The values are derived via (t-t_min)/(t_max-t_min), t_min=-16, t_max=+50 (only in hourly scale)

hum: Normalized humidity. The values are divided to 100 (max)

windspeed: Normalized wind speed. The values are divided to 67 (max)

casual: count of casual users

registered: count of registered users

cnt: count of total rental bikes including both casual and registered

for further more information please go through the following link,
http://archive.ics.uci.edu/ml/datasets/Bike+Sharing+Dataset

## Visualization & Explanatory Analysis

```python
# Visualizing monthly counts across the seasons
fig,ax=plt.subplots(figsize=(16,8))
sns.pointplot(data=bike_df[['mnth','cnt','season']],
              x='mnth',
              y='cnt',
              hue='season',
              ax=ax)
ax.set(title='Season wise monthly distribution of rental counts')
```


Distribusi di atas menggambarkan pola jumlah pengguna sepanjang musim. Jumlah pengguna **paling rendah** adalah selama ***musim Semi***, diikuti oleh peningkatan bertahap selama ***musim Panas***. Selama ***musim Gugur***, jumlah pengguna tetap **relatif stabil**. Jumlah tersebut kemudian **mencapai puncak** pada awal ***musim Dingin*** namun kemudian mengalami penurunan


```python
# Visualizing monthly counts across different weathers
fig, ax = plt.subplots(figsize=(16, 8))
sns.pointplot(data=bike_df,
              x='hr',
              y='cnt',
              hue='weathersit',
              ax=ax)
ax.set(title='Weathers wise monthly distribution of rental counts')
```

Dari distribusi di atas, dapat dilihat bahwa jumlah pengguna sepeda paling tinggi pada saat cuaca cerah (1) dan menyusul di bawahnya pada saat cuaca berawan (2). Sedangkan pada saat hujan dan salju ringan (3), jumlah pengguna berkurang. Saat badai hujan dan salju (4), pengguna hanya sedikit saja pada jam tertentu.

```python
# Visualizing weekday and weekend counts
fig,ax=plt.subplots(figsize=(16, 8))
sns.pointplot(data=bike_df, x='hr', y='cnt', hue='weekday', ax=ax)
ax.set(title='Count of bikes during weekdays and weekends')
```

Distribusi di atas menggambarkan pola pengguna sepeda pada hari-hari dalam seminggu. Terlihat bahwa pengguna meningkat pada weekend (0 dan 6) lebih tinggi pada pukul 9 sampai 16. Sedangkan pada weekday (1-5) lebih tinggi pada pukul 5 sampai 9 dan pukul 16 sampai 23.

```python
# Visualizing monthly counts
fig,ax=plt.subplots(figsize=(16,8))
sns.barplot(data=bike_df[['mnth','cnt']],x='mnth',y='cnt',ax=ax)
ax.set(title='Month-wise rental distribution')
```

Bar Plot di atas menampilkan jumlah pengguna stabil naik dari bulan Januari (1) sampai pada puncaknya di bulan Juni (6), kemudian stabil sampai bulan September (9) dan perlahan turun sampai bulan Desember (12)

```python
# Visualizing holiday vs workday counts
bike_df['category'] = bike_df['holiday'].apply(lambda x: 'holiday' if x else 'workingday')
bike_df = bike_df.groupby(['category','yr']).agg({
        'cnt': 'sum'
    }).reset_index()
sns.set_style('whitegrid')
fig,ax=plt.subplots(figsize=(16, 8))
sns.barplot(x='yr', y='cnt', hue='category', data=bike_df)
ax.set(title='Count of bikes during workingday vs holiday')
plt.show()
```

```python
# Visualizing Year count
fig, ax = plt.subplots(figsize=(16, 8))
sns.violinplot(data=bike_df[['yr', 'cnt']], x='yr', y='cnt', ax=ax)
ax.set(title='Year-wise distribution of rental counts')
plt.show()
```

tahun 0 = 2011, tahun 1 = 2012

Distribusi di atas dengan jelas membantu kita memahami distribusi dalam jumlah pengguna pada tahun 2011 dan 2012. Distribusi untuk tahun 2012 memiliki puncak pada nilai tertinggi dibandingkan dengan distribusi untuk tahun 2011.

```python
#Plotting pairplot of the numeric variables

sns.pairplot(bike_df[["temp","atemp","hum","windspeed","cnt"]])
plt.figure(figsize=(2, 2))
plt.show()
```

```python
# Plotting correlation of all variables using heatmap
# kecuali "casual" dan "registered" karena kedua variable telah di-agregasi dalam variable "cnt"

plt.figure(figsize = (12,10))
sns.heatmap(bike_df[["temp","atemp","hum","windspeed","season","hr","yr","mnth","holiday","weekday","workingday","weathersit","cnt"]].corr(), annot = True, cmap="Blues")
plt.show()
```

Dari Heatmap di atas, dapat diketahui bahwa

variable yang mempunyai korelasi positif tertinggi dengan jumlah permintaan sewa sepeda adalah "temp" dan "atemp" --- yang berarti temperatur mempengaruhi kenaikan permintaan sewa.

varible yang mempunyai korelasi negatif tertinggi dengan jumlah permintaan sewa sepeda adalah "hum" - yang berarti semakin tinggi kelembaban udara, semakin menurun permintaan sewa.
